```python
%sql
/*10 GROUP BY Practice Questions 

Shape 

Q1 Find the total sales amount for each region. 

Q2 Find total sales amount for each product. 

Q3 Find number of orders in each region. 

Q4 Find average sales amount in each region. 

Q5 Find highest sales amount in each region. 

Q6 Find lowest sales amount in each product category. 

Q7 Find total sales by region and product. 

Q8 Find number of orders by product and region. 

Q9 Find average sales amount for each product. 

Q10 Find regions where total sales > 3,00,000. 
*/
```

```python
%sql
--Q1 Find the total sales amount for each region. 
Select region, sum(amount) as total_amount_sales
from ecommerceonline.shopping.sales_orders
group by region;
```

```python
%sql
--Q2 Find total sales amount for each product. 
select product, sum(amount) as Total_product_sales
from ecommerceonline.shopping.sales_orders
group by product;
```

```python
%sql
--Q3 Find number of orders in each region. 
select region, count(order_id) as numbers_of_orders
from ecommerceonline.shopping.sales_orders
group by region;
```

```python
%sql
--Q4 Find average sales amount in each region. 
select region, avg(amount) as Avg_amount_sale
from ecommerceonline.shopping.sales_orders
group by region;
```

```python
%sql
--Q5 Find highest sales amount in each region.
select region, max(amount) as highest_sales_amount
from ecommerceonline.shopping.sales_orders
group by region;
--row number() over (partetion region group by order) 
```

```python
%sql
--Q6 Find lowest sales amount in each product category. 
select product,min(amount)as lowest_sales_amount
from ecommerceonline.shopping.sales_orders
group by product;
```

```python
%sql
--Q7 Find total sales by region and product.
select region,product, sum(amount) as total_region_product_sales
from ecommerceonline.shopping.sales_orders
group by region,product;
```

```python
%sql
--Q8 Find number of orders by product and region. 
select region,product, count(order_id)as number_of_orders
from ecommerceonline.shopping.sales_orders
group by region,product;
 
```

```python
%sql
--Q9 Find average sales amount for each product. 
select product,avg(amount)as avg_sales_amount
from ecommerceonline.shopping.sales_orders
group by product;
 
```

```python
%sql
--Q10 Find regions where total sales > 3,00,000.
SELECT region, SUM(amount) AS total_sales
FROM ecommerceonline.shopping.sales_orders
GROUP BY region
HAVING SUM(amount) >=300000;
 
```
