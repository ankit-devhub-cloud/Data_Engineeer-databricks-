### Find all customers who have not placed any orders.

Customers
id name 
- 1  aa
- 2  bb
- 3  cc
- 4  dd
- 5  ee

Orders
id cust_id
- 1  2
- 2  2
- 3  3
- 4  3
- 5  5


```python
customers = [(1,'aa'),(2,'bb'),(3,'cc'),(4,'dd'),(5,'ee')]
orders = [(1,2),(2,2),(3,3),(4,3),(5,5)]

df_customers = spark.createDataFrame(customers,['id','name'])
df_orders = spark.createDataFrame(orders,['id','cust_id'])
```

###customers who placed orders with inner join

```python
df_cust_placed_orders = df_customers.join(df_orders,df_customers.id == df_orders.cust_id,'inner').select(df_customers.id,df_customers.name).distinct()
display(df_cust_placed_orders)
```

##customers who placed order and not placed order with left join

```python
df_left_join = df_customers.join(df_orders,df_customers.id == df_orders.cust_id,'left')
display(df_left_join)
```

```python
df_customers_placed_orders = df_customers.join(df_orders,df_customers.id == df_orders.cust_id,'left').filter(df_orders.id.isNotNull()).distinct()
display(df_customers_placed_orders)
```

```python
df_cust_not_placed_orders = df_customers.join(df_orders,df_customers.id == df_orders.cust_id,'left').filter(df_orders.id.isNull())
display(df_cust_not_placed_orders)
```
