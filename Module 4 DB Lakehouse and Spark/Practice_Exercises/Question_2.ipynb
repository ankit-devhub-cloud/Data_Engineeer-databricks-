### List all products that are currently out of stock.

```python
products =  [(1, "Product1"), (2, "Product2"), (3, "Product3"), (4, "Product4"), (5, "Product5")]
stock =[(1, 100,1), (2, 200,3), (3, 300,3)]
```

```python
df_products = spark.createDataFrame(products,["id","name"])
df_products.createOrReplaceTempView("products")
df_stock = spark.createDataFrame(stock,["id","stock","product_id"])
df_stock.createOrReplaceTempView("stocks")
```

Again we can use the left join

```python
df_prod_stock = df_products.join(df_stock,df_products.id == df_stock.product_id,'left')
display(df_prod_stock)
```

```python
%sql
select * from products
```

```python
%sql
select * from stocks
```

```python
%sql

select products.id,products.name,stocks.product_id sp,*
from products left join stocks on products.id = stocks.product_id
where stocks.product_id is null
```

```python
df_prod_out_of_stock = df_products.join(df_stock,df_products.id == df_stock.product_id,'left').filter(df_stock.product_id.isNull())
display(df_prod_out_of_stock)
```

The above results are products where we did not find any entry in stocks table. which implies. This implies stocks is empty for those products.
