### Find the total number of employees working in each department.

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

employees = [
  {"id": 1, "name": "John", "department_id": 1},
  {"id": 2, "name": "Jane", "department_id": 2},
  {"id": 3, "name": "Doe", "department_id": 3},
  {"id": 4, "name": "Tom", "department_id": 2}
  ]

department = [
  {"id": 1, "name": "Sales"},
  {"id": 2, "name": "Marketing"}
  ]

# Schema for employees
employee_schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("department_id", IntegerType(), True)
])

# Schema for department
department_schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True)
])


employees_df =  spark.createDataFrame(employees,schema=employee_schema)
employees_df.createOrReplaceTempView("employees")
department_df = spark.createDataFrame(department,schema=department_schema)
department_df.createOrReplaceTempView("department")
```

```python
%sql

select e.id as emp_id, e.name as emp_name, d.name as dept_name
from employees as e 
left join department as d on e.department_id = d.id
```

```python
%sql
select d.name as dept_name, count(*)
from employees as e 
left join department as d on e.department_id = d.id
where d.name is not null
group by d.name
```
