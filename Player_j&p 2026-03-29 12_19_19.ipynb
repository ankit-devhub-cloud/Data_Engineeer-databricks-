```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder.appName("Exercise6").getOrCreate()
df = spark.read.csv("/Volumes/ecommerceonline/shopping/orders_1/orders (2).csv", header=True, inferSchema=True)
df2 = df.withColumn("total_amount", col("quantity") * col("price"))
df_laptop = df2.where(col('product') == 'Laptop')
df_laptop.show()
```
